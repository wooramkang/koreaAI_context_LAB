## 1. Sigmoid function

Currently, sigmoid function does not used in deep learning. Because of the Vanishing Gradient problem.
But, until we discovered this problem, it was often used in deep learning.

So, i recommend to learn about sigmoid function.

### Let's draw the sigmoid function.



$$
\sigma(z) = {1 \over {1+ e^{-z}}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    
    output = []
    
    for y in x:
        output.append(1 / (1 + np.exp(-y)))
    return output

x = np.linspace(-20, 20, 100)

sig = sigmoid(x)

plt.plot(x, sig)
plt.grid()
plt.show()
```

### Now let's draw the derivative function of this.

In this graph, you can see the gradient is almost zero when the sigmoid function's value is 0 or 1.

In the process of the back propagation, almost zero value is multiplied on the weight.

So, the weight is not updated or updated very slowly.

In addition, due to the exponential function, the exponential function remains even if the derivative is used.
This result will make the increasment of the computation.

Lastly, the center value of the graph is not zero, it causes a convergence speed decrease.

```python
def de_sigmoid(x):
    
    output = []
    
    for y in x:
        output.append(np.exp(-y) / (1 + np.exp(-y))**2)
    return output

x = np.linspace(-20, 20, 100)

de_sig = de_sigmoid(x)

plt.plot(x, de_sig)
plt.grid()
plt.show()
```
